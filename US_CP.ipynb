```
#pip install pandas_bokeh


import numpy as np
import pandas_bokeh
import pandas as pd
import matplotlib.pyplot as plt

pandas_bokeh.output_notebook()

tables = pd.read_html("https://www.federalreserve.gov/releases/cp/rates.htm")

CP = tables[-1]

CP_Daily = CP.iloc[18:23]
CP_Daily = CP_Daily.iloc[:,:7]
CP_Daily_clean = CP_Daily.replace(to_replace ="n.a.", value ="0") 

#CP_Daily['Period'] = pd.to_datetime(CP_Daily['Period'])
CP_Daily_clean = CP_Daily_clean.set_index('Period')
CP_Daily_clean = CP_Daily_clean.astype('float64')
print(CP_Daily_clean.head(5))



fig = plt.figure(figsize=(20,10))
CP_Daily_clean.plot.bar()
plt.ylabel('CP Avg. Rate', fontsize = 12) #for y label
plt.xlabel('Date', fontsize = 12) #for x label
plt.suptitle("Fed - US CP data", fontsize=15, fontweight=0, color='black')
plt.tight_layout()
plt.show()

tables = pd.read_html("https://www.federalreserve.gov/releases/cp/volumestats.htm")

CPV = tables[2]
CPV = CPV.iloc[18:23]
CPV = CPV.iloc[:,[0,3,4,5,6,7]]
CPV_Daily = CPV.replace(to_replace ="n.a.", value ="0") 
CPV_Daily = CPV_Daily.set_index('Period')
print(CPV_Daily)

CPV_clean = CPV_Daily.astype('float64')
CPV_clean.plot.bar()
```
