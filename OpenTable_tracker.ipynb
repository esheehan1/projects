```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly
import plotly.graph_objs as go
import plotly.express as px
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
cf.go_offline()
cf.set_config_file(offline=False, world_readable=True)
cf.set_config_file(theme='white')
init_notebook_mode(connected=True)

# Data Source: https://www.opentable.com/state-of-industry

url = 'https://raw.githubusercontent.com/esheehan1/projects/master/YoY_Seated_Diner_Data.csv'
input_data = pd.read_csv(url)
print(input_data.head())
```

```
countries_data = input_data[input_data['Type'] == 'country'].drop(columns=['Type']).set_index('Name').T
countries_data
print(countries_data.head)

countries_data.plot()

plt.grid(True)
plt.legend()
plt.ylabel('YoY % Change', fontsize = 12) #for y label
plt.xlabel('Date', fontsize = 12) #for x label
plt.suptitle("Global Restaurant Reservations", fontsize=15, fontweight=0, color='black')
plt.tight_layout()
plt.show()
```
